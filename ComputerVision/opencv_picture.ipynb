### OpenCV(Computer Vision)

다양한 영상(이미지)/동영상 처리에 사용되는 오픈소스

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/MyDrive/opencv/image1.jpg') # 해당 경로의 이미지 입력
img_gray = cv2.imread('/content/drive/MyDrive/opencv/image1.jpg', cv2.IMREAD_GRAYSCALE)
img_color = cv2.imread("/content/drive/MyDrive/opencv/image1.jpg", cv2.IMREAD_COLOR)
img_unchanged = cv2.imread('/content/drive/MyDrive/opencv/image1.jpg', cv2.IMREAD_UNCHANGED)

# cv2_imshow(img_gray) # img 변수를 창에 보여줌
# cv2_imshow(img_color)
cv2_imshow(img_unchanged)

key = cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 읽기옵션

1. cv2.IMREAD_COLOR : 컬러 이미지. 투명 영역은 무시(기본값)
2. cv2.IMREAD_GRAYSCALE : 흑백 이미지.
3. cv2.IMREAD_UNCHANGED : 투명 영역도 포함
