# 빈 스케치북 만들기

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# 세로 480 x 가로 640, 3 channels (RGB)에 해당하는 스케치북 생성
img = np.zeros((480, 640, 3), dtype=np.uint8)

# img[:] = (255, 0, 0) # B,G,R // black(0, 0, 0), white(255, 255, 255)
img[100:200, 200:300] = (255,255,255)

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 직선 그리기

1. cv2.LINE_4 : 상하좌우 4방향으로 연결된 선
1. cv2.LINE_8 : 대각선까지 포함된 8방향으로 연결된 선
1. cv2.LINE_AA : 부드러운 선(anti-aliasing)

```python
img_ = np.zeros((480,640,3), dtype = np.uint8)

COLOR = (0,255,255) # BGR :: 노랑색...
THICKNESS = 3 # 선 두께

# 그릴 위치, 시작점, 끝점, 색깔, 두께, 선종류
cv2.line(img_,(50,100), (400,50), COLOR, THICKNESS, cv2.LINE_4)
cv2.line(img_,(150,100), (500,50), COLOR, THICKNESS, cv2.LINE_8)
cv2.line(img_,(250,150), (600,100), COLOR, THICKNESS, cv2.LINE_AA)


cv2_imshow(img_)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 원 그리기

```python
_img = np.zeros((480,640,3), dtype = np.uint8)

COLOR = (255,255,0) # BGR : 옥색
RADIUS = 50 # 반지름 값
THICKNESS = 3 # 두께

# 스케치북 img, 원의 중심, 반지름, 색깔, 두께, 선종류
cv2.circle(_img, (200,100), RADIUS, COLOR, THICKNESS, cv2.LINE_AA)
cv2.circle(_img, (400,100), RADIUS, COLOR, cv2.FILLED, cv2.LINE_AA)

cv2_imshow(_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

### 사각형 그리기

```python
_img_ = np.zeros((480,640,3), dtype = np.uint8)

COLOR = (100,255,100)
THICKNESS = 3

# 스케치북_img, 왼쪽 위의 좌표, 오른쪽 아래 좌표, 색깔, 두께
cv2.rectangle(_img_, (100, 100), (200, 200), COLOR, THICKNESS)
cv2.rectangle(_img_, (300, 100), (400, 300), COLOR, cv2.FILLED)

cv2_imshow(_img_)
cv2.waitKey()
cv2.destroyAllWindows()
```

### 파일저장

```python
fill_rectangle = cv2.rectangle(_img_, (300, 100), (400, 300), COLOR, cv2.FILLED)
cv2.imwrite('rectangle.jpg',fill_rectangle)
```

### 다각형 그리기

```python
img123 = np.zeros((480,640,3), dtype = np.uint8)

COLOR = (0,0,255) # BGR :RED 빨강
THICKNESS = 3 # 두께

pts1 = np.array([[100,100],[200,100],[100,200]])
pts2 = np.array([[200,100],[300,100],[300,200]])

# cv2.polylines(img123, [pts1], True, COLOR, THICKNESS, cv2.LINE_AA)
# cv2.polylines(img123, [pts2], True, COLOR, THICKNESS, cv2.LINE_AA)

cv2.polylines(img123, [pts1, pts2], True, COLOR, THICKNESS, cv2.LINE_AA)

## 꽉 찬 2개의 대칭 삼각형을 한 번 이 밑에 그려봐주세요.
pts3 = np.array([[100,300],[200,300],[100,400]])
pts4 = np.array([[200,300],[300,300],[300,400]])

cv2.fillPoly(img123, [pts3, pts4], COLOR, cv2.LINE_AA) # 꽉 찬 삼각형

cv2_imshow(img123)
cv2.waitKey()
cv2.destroyAllWindows()
```
